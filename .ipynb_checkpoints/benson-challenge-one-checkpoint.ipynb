```python
import csv
with open("turnstile_150404.txt", "r") as turn_data:
    reader=csv.reader(turn_data)
    my_hash={}
    for row in reader:
        key = row[0:4]
        val = row[4:-1]
        if my_hash[key] != None:
            my_hash[key].append( [ val ] )
        else:
            my_hash[key] = [ [ val ] ] 
print my_hash
```
